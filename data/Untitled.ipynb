```python
import matplotlib.pyplot as plt
import numpy as np
import re
from plotstyles import tools
tools.set_mpl_rcParams()
```

```python
x,y = [],[]
with open('./mesh2d.dat') as file:
    for i,line in enumerate(file):
        if i > 1:
            _, str1, str2, _ = re.split('\s+', line)  
            x.append(float(str1))
            y.append(float(str2))
x = np.array(x)
y = np.array(y)
z = np.zeros_like(x)
```

```python
fig = plt.figure(figsize=tools.cm2inch([12,8]), dpi=500)
ax = fig.add_subplot(111)
ax.scatter(x, y, s=0.5)
ax.set_aspect(1)

```
